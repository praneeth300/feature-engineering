##Random Sample Imputation
Aim: Random sample imputation consists of taking random observation from the dataset and we use this observation to replace the nan values

When should it be used? It assumes that the data are missing completely at random(MCAR)

```python
data=pd.read_csv("../input/titanic/train.csv",usecols=['Survived','Fare','Age'])
data.head()
```

```python
data.isnull().mean()
```

```python
data.isnull().sum()
```

```python
data['Age'].dropna().sample(data['Age'].isnull().sum(),random_state=0)
```

```python
data[data['Age'].isnull()].index
```

```python
def impute_nan(data,variable,median):
    data[variable+'_median']=data[variable].fillna(median)
    data[variable+'_random']=data[variable]
    random_sample=data[variable].dropna().sample(data[variable].isnull().sum(),random_state=0)
    random_sample.index=data[data[variable].isnull()].index
    data.loc[data[variable].isnull(),variable+'_random']=random_sample
```

```python
medain=data.Age.median()
median
```

```python
impute_nan(data,'Age',median)
```

```python
data.head()
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
data.Age.plot(kind='kde',ax=ax,color='yellow')
data.Age_median.plot(kind='kde',ax=ax,color='green')
data.Age_random.plot(kind='kde',ax=ax,color='blue')
lines, labels=ax.get_legend_handles_labels()
ax.legend(lines,labels,loc='best')
```

### End of distribution imputaion


```python
data=pd.read_csv("../input/titanic/train.csv",usecols=['Survived','Fare','Age'])
data.head()
```

```python
data.Age.hist(bins=50)
```

```python
extreme=data.Age.mean()+3*data.Age.std()
```

```python
extreme
```

```python
sns.boxplot('Age',data=data)
```

```python
def impute_nan(data,variable,median,extreme):
    data[variable+'_end_distribution']=data[variable].fillna(extreme)
    data[variable].fillna(median,inplace=True)
```

```python
impute_nan(data,'Age',data.Age.median(),extreme)
```

```python
data.head()
```

### Arbitrary value imputation


```python
def impute_nan(data,variable):
    data[variable+'_zero']=data[variable].fillna(0)
    data[variable+'_hundred']=data[variable].fillna(100)
```

```python
impute_nan(data,'Age')
```
